```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('/projects/DEIKE/jiarongw/jiarongw-postprocessing/media/matplotlib/stylelib/thesis.mplstyle')
```

```python
means = []
for i in range(2,33):
    df = pd.read_csv('./data/Grare2018/umean_%g.txt' %i, delimiter=' ')
    df_ = df.dropna().reset_index(drop=True)
    means.append(df_)
```

```python
fig = plt.figure(figsize=[2,2])
for i,df in enumerate(means):
    print(df.U[0], df.U.max())
    plt.plot(np.array(df.U), df.z, c=plt.get_cmap('Greens')(df.U[0]/10))
plt.ylim([0, 300])
plt.xlim([0, 12])
plt.xlabel(r'$\overline{u} \mathrm{\;(m/s)}$')
plt.ylabel(r'$z \mathrm{\;(m)}$')
# plt.yscale('log')
```

```python
fig = plt.figure(figsize=[2,2])
for i,df in enumerate(means):
    print(df.U[0], df.U.max())
    plt.plot(np.array(df.U), df.z, c=plt.get_cmap('Greens')(df.U[0]/10))
plt.ylim([1, 500])
plt.xlim([0, 12])
plt.yscale('log')
plt.xlabel(r'$\overline{u} \mathrm{\;(m/s)}$')
plt.ylabel(r'$z \mathrm{\;(m)}$')
```

```python
fig = plt.figure(figsize=[2,2])
for i in (5,6,8,9,10):
    print(means[i].U[0], means[i].U.max())
    plt.plot(np.array(means[i].U), means[i].z, '-o', c=plt.get_cmap('Greens')(means[i].U[0]/10))
plt.ylim([1, 300])
plt.xlim([0, 12])
plt.xlabel(r'$\overline{u}$')
plt.ylabel(r'$z$')
plt.yscale('log')
```

```python
t1 = (6,26,66,86)
t2 = (12,52,92,132)
t3 = (24,56,88,120)
mean_CU2 = []
for t in t1:
    mean_simu = pd.read_csv('../turbulence/shared_data/mean_uair_CU2/omegat%g.txt' %t, 
                delimiter=',', header=0, names=['z','U'])
    mean_CU2.append(mean_simu)
```

```python
fig = plt.figure(figsize=[1,2])
ax = plt.gca()
ax2 = ax.twinx()
for i, df in enumerate(mean_CU2):
    ax.plot(np.array(df.U)[:-5], df.z[:-5], '-o', c=plt.get_cmap('Greens')((i+1)/5))
    ax2.plot(np.array(df.U)[:-5], df.z[:-5]/720, '-o', c=plt.get_cmap('Greens')((i+1)/5))
ax.set_ylim([1, 720])
ax2.set_ylim([1/720, 1])
ax.set_xlim([0, 25])
ax.set_xlabel(r'$\overline{u}/u_*$')
ax.set_ylabel(r'$zu_*/\nu$')
ax2.set_ylabel(r'$z/H$')
ax.set_yscale('log'); ax2.set_yscale('log')
ax.set_xticks([0,5,10,15,20,25])
plt.show()
```
